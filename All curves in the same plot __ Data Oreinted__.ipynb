```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
# sns.set()
```

```python
z  = np.linspace(0, 0.5, 1000)
```

**Reading the data** 
- data_1: flowrate = 0.1 "b"
- data_2: flowarate = 0.3 "g"
- data_3: flowaret = 0.45 "y"
- data_4: flowrate = 0.6 "m"

```python
data_1 = pd.read_csv("flowrate_0.1_.csv")
data_2 = pd.read_csv("flowrate_0.3_.csv")
data_3 = pd.read_csv("flowrate_0.45_.csv")
data_4 = pd.read_csv("flowrate_0.6_.csv")
data_1.head() #just to realize the header
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(3, 3, 1)
plt.xlabel("channel length (m)", fontsize=15)
plt.ylabel("Conversion of CO2%", fontsize=15)
plt.title("Conversion of CO2 vs. Channel Length", fontweight = "bold")
plt.xlim(0,0.5)
plt.ylim(0,100)
plt.plot(z,data_1["CO2 conversion"],'b')
plt.plot(z,data_2["CO2 conversion"],'g')
plt.plot(z,data_3["CO2 conversion"],'y')
plt.plot(z,data_4["CO2 conversion"],'m')




plt.subplot(3, 3, 2)
plt.xlabel("channel length (m)", fontsize=15)
plt.ylabel("Conversion of MEA%", fontsize=15)
plt.title("Conversion of MEA vs. Channel Length", fontweight = "bold")
plt.xlim(0,0.5)
plt.ylim(0,100)
plt.plot(z,data_1["MEA conversion"],'b')
plt.plot(z,data_2["MEA conversion"],'g')
plt.plot(z,data_3["MEA conversion"],'y')
plt.plot(z,data_4["MEA conversion"],'m')


plt.subplot(3, 3, 3)
plt.xlabel("Liquid Flowa rate ml/min", fontsize=15)
plt.ylabel("$k_{GL}a_{GL}$ 1/s", fontsize=15)
plt.title("Change of Mass Transfer Coefficient", fontweight = "bold")
plt.xlim(0,1)
plt.ylim(0,1.5)
plt.plot(0.1,0.0028, 'bx')
plt.plot(0.3,0.0459, 'gx')
plt.plot(0.45,0.1265, 'yx')
plt.plot(0.6,0.27, 'mx')


plt.subplot(3, 3, 4)
plt.xlabel("channel length (m)", fontsize=15)
plt.ylabel("Concentration of CO2 gas (mol/m3)", fontsize=15)
plt.title("Concentration of CO2 gas vs. Channel Length", fontweight = "bold")
plt.xlim(0,0.5)
plt.ylim(0,40)
plt.plot(z,data_1["CO2G"],'b')
plt.plot(z,data_2["CO2G"],'g')
plt.plot(z,data_3["CO2G"],'y')
plt.plot(z,data_4["CO2G"],'m')

plt.subplot(3, 3, 5)
plt.xlabel("channel length (m)", fontsize=15)
plt.ylabel("Concentration of MEA (mol/m3)", fontsize=15)
plt.title("Concentration of MEA vs. Channel Length", fontweight = "bold")
plt.xlim(0,0.5)
plt.ylim(0,1000)
plt.plot(z,data_1["MEA"],'b')
plt.plot(z,data_2["MEA"],'g')
plt.plot(z,data_3["MEA"],'y')
plt.plot(z,data_4["MEA"],'m')

plt.subplot(3, 3, 6)
plt.xlabel("Liquid Flowa rate ml/min", fontsize=15)
plt.ylabel("CO2 conversion %", fontsize=15)
plt.title("Change of Mass Transfer Coefficient", fontweight = "bold")
plt.xlim(0,1)
plt.ylim(0,100)
LiqFlowrate_exp=[0.1,0.3,0.45,0.6]      # from Dr. mamoun Lab
Conversion_exp=[0.97,27.66,56.64,89.13] #  from Dr. mamoun Lab
plt.plot(LiqFlowrate_exp,Conversion_exp,"rx",label="experiment")
plt.plot(0.1,data_1["CO2 conversion"][999],'bx', label ="0.1 ml/min")
plt.plot(0.3,data_2["CO2 conversion"][999],'gx', label ="0.3 ml/min")
plt.plot(0.45,data_3["CO2 conversion"][999],'yx', label ="0.45 ml/min")
plt.plot(0.6,data_4["CO2 conversion"][999],'mx', label ="0.6 ml/min")
legend = plt.legend(loc='lower right', shadow=True, fontsize='large')



plt.tight_layout()
plt.show()
```
